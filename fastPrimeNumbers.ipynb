```python
import random

def is_prime(n, k=5):
    """Miller-Rabin primality test."""
    if n <= 1:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0:
        return False

    # Write n as d*2^r + 1
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    # Perform the Miller-Rabin test k times
    for _ in range(k):
        a = random.randint(2, n - 2)
        x = pow(a, d, n)
        if x == 1 or x == n - 1:
            continue
        for _ in range(r - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False  # n is composite

    return True  # n is probably prime

# Example usage:
number_to_check = 137
if is_prime(number_to_check):
    print(f"{number_to_check} is prime.")
else:
    print(f"{number_to_check} is not prime.")
```

```python
import random

def pollards_rho(n):
    if n % 2 == 0:
        return 2

    x = random.randint(1, n-1)
    y = x
    c = random.randint(1, n-1)
    d = 1

    f = lambda x: (x**2 + c) % n

    while d == 1:
        x = f(x)
        y = f(f(y))
        d = math.gcd(abs(x-y), n)

    return d

def prime_factors(n):
    factors = []
    while n > 1:
        factor = pollards_rho(n)
        factors.append(factor)
        n //= factor
    return factors

# Example usage:
number_to_factorize = 148293123
factors = prime_factors(number_to_factorize)
print(f"Prime factors of {number_to_factorize}: {factors}")
```

```python
def prime_factors(n):
    factors = []
    # Handle divisible by 2
    while n % 2 == 0:
        factors.append(2)
        n //= 2

    # Handle odd factors starting from 3
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n //= i

    # If n is a prime number greater than 2
    if n > 2:
        factors.append(n)

    return factors

# Example usage:
number_to_factorize = 150
result = prime_factors(number_to_factorize)
print(f"Prime factors of {number_to_factorize}: {result}")
```

```python
import math
import random

def prime_factors_random(num):
    array_list = []
    trials = set()

    while len(array_list) < 10:
        if num // 2 <= 2:
            return array_list
        x = random.randint(2, num // 2)
        trials.clear()

        for _ in range(10 * (len(array_list) + num) ** 2):
            a = num / x

            if 2 > abs(a - math.floor(a)) / 0.5 > 0.5:
                x += num // int(a+1) ** 10
            else:
                x -= num // int(a+1) ** 10

            if int(x) == x:
                trials.add(int(x))

            if num % x == 0:
                if x not in array_list:
                    array_list.append(x)
                    num //= x
                    break

    return array_list

# Example usage:
number = 60
result = prime_factors_random(number)
print(f"Prime factors of {number}: {result}")

    
```

```python
def factorize_native(num):
    factors = []
    divisor = 2

    while num > 1:
        while num % divisor == 0:
            factors.append(divisor)
            num //= divisor
        divisor += 1

    return factors

# Example usage:
num_to_factorize = 12371233 * 14828313
result = factorize_native(num_to_factorize)
print(f"Factors of {num_to_factorize}: {result}")
```
