```python
def is_lychrel(number, max_iterations=50):
    for _ in range(max_iterations):
        # Reverse the digits and add to the original number
        number += int(str(number)[::-1])
        
        # Check if the result is a palindrome
        if str(number) == str(number)[::-1]:
            return False
    
    # If no palindrome is found within the iterations, consider it a potential Lychrel number
    return True

# Example usage:
number_to_check = 1984
result = is_lychrel(number_to_check)

if result:
    print(f"{number_to_check} is a potential Lychrel number.")
else:
    print(f"{number_to_check} is not a Lychrel number.")
```
